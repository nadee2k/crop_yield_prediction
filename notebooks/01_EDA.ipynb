```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load CSVs
yield_df = pd.read_csv('../data/crop_yield.csv')
weather_df = pd.read_csv('../data/weather.csv')
soil_df = pd.read_csv('../data/soil.csv')
```

```python
# Merge
df = pd.merge(yield_df, weather_df, on=['location', 'year'])
df = pd.merge(df, soil_df, on='location')

# Quick look
df.head()

# Summary
df.describe()

# Check nulls
df.isnull().sum()

```

```python
# Histogram of yield
sns.histplot(df['yield (kg/hectare)'], kde=True)
plt.title('Distribution of Tea Yield')
plt.show()
```

```python

# Correlation heatmap
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python

# Boxplot of yield by location
sns.boxplot(x='location', y='yield (kg/hectare)', data=df)
plt.title('Yield by Location')
plt.show()
```

```python

# Boxplot of yield by variety
sns.boxplot(x='variety', y='yield (kg/hectare)', data=df)
plt.title('Yield by Variety')
plt.show()
```
